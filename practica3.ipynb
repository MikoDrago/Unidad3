#Practica 3: Simular los tiempos de servicio o inactividad en una f√°brica de chips.

##M√©todo de la transformada inversa:

Este m√©todo transforma n√∫meros aleatorios uniformes (0,1) en variables de una distribuci√≥n espec√≠fica.

##Distribuci√≥n Lognormal:

Se utiliza principalmente para modelar tiempos de servicio o inactividad con asimetr√≠a positiva. Los valores son siempre positivos y tienden a distribuirse asim√©tricamente con una "cola" hacia la derecha, usada cuando los tiempos de servicio tienen gran variabilidad.

##Ejemplo de Aplicaci√≥n:

En una f√°brica de chips, los tiempos de inactividad de brazos automatizados pueden seguir una distribuci√≥n lognormal, permitiendo simular estos tiempos y planificar mantenimientos preventivos.

##Pasos para aplicar el m√©todo:

Generar un n√∫mero aleatorio

Aplicar la transformada inversa usando la inversa de la funci√≥n de distribuci√≥n acumulada est√°ndar normal para obtener la variable aleatoria lognormal.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mu = 1.0
sigma = 0.5
n = 100


U = np.random.uniform(0, 1, n)

Z = mu + sigma * stats.norm.ppf(U)
T = np.exp(Z)

print("Tiempos de inactividad generados:", T)

plt.hist(T, bins=20, edgecolor='black', density=True)

x = np.linspace(min(T), max(T), 100)
pdf = stats.lognorm.pdf(x, sigma, scale=np.exp(mu))
plt.plot(x, pdf, 'r', label='Distribuci√≥n lognormal te√≥rica')

plt.title("Histograma de los tiempos de inactividad (Distribuci√≥n Lognormal)")
plt.xlabel("Tiempo de inactividad")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.show()
```
