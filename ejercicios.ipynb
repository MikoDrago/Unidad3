#Problema 1:
#Distribución Binomial
##Contexto del Problema
Un distribuidor de artículos electrónicos recibe diariamente múltiples pedidos, pero cuenta con un inventario limitado, lo que le impone una probabilidad del 70% de poder satisfacer cada solicitud. La empresa desea anticipar cuántos pedidos diarios puede satisfacer para optimizar su gestión de inventario y minimizar los faltantes de stock.

##Objetivo Específico
Simular, usando una variable aleatoria binomial, el número de pedidos satisfechos en un día a lo largo de una semana, con 20 pedidos cada día.

##Distribución Utilizada
La Distribución Binomial es ideal para modelar un evento con dos posibles resultados: éxito o fracaso (en este caso, satisfacer o no un pedido). Se caracteriza por dos parámetros: el número de ensayos (20 pedidos diarios) y la probabilidad de éxito (70%).

```python
import numpy as np
import matplotlib.pyplot as plt

dias = 7
pedidos_por_dia = 20
probabilidad_exito = 0.7

satisfaccion_pedidos = np.random.binomial(pedidos_por_dia, probabilidad_exito, dias)

media = np.mean(satisfaccion_pedidos)
desviacion_estandar = np.std(satisfaccion_pedidos)

plt.hist(satisfaccion_pedidos, bins=range(15, 21), color='skyblue', edgecolor='black')
plt.title("Distribución de Pedidos Satisfechos")
plt.xlabel("Pedidos Satisfechos")
plt.ylabel("Frecuencia")
plt.show()

print(f"Media: {media}, Desviación Estándar: {desviacion_estandar}")


print("Tiempos de espera en el mostrador de servicio técnico:", satisfaccion_pedidos)
```

##Conclucion Del resultado
El arreglo satisfaccion_pedidos proporciona un registro diario de la cantidad de pedidos satisfechos a lo largo de una semana. Este seguimiento es crucial para evaluar el desempeño del inventario, ya que permite identificar patrones y detectar días en los que podría haber faltantes. Con esta información, se pueden tomar decisiones informadas para mejorar la gestión del inventario y asegurar que la demanda de los clientes se satisfaga de manera eficiente.

#Problema 2:
# Distribución LogNormal
##Contexto del Problema
En el mostrador de servicio técnico de una tienda departamental, los tiempos de espera de los clientes presentan una gran variabilidad, con un promedio de espera de 10 minutos y una desviación estándar de 2 minutos. Este comportamiento se ajusta a una distribución lognormal, útil para describir fenómenos de espera.

##Objetivo Específico
Simular los tiempos de espera de 50 clientes para estudiar el comportamiento de la fila y la duración promedio.

##Distribución Utilizada
La Distribución LogNormal es adecuada para modelar variables cuya transformación logarítmica sigue una distribución normal, como tiempos de espera con una variabilidad que crece en proporción a la media.

```python
clientes = 50
media = 10
desviacion_estandar = 2

tiempos_espera = np.random.lognormal(mean=np.log(media), sigma=desviacion_estandar, size=clientes)
print("Tiempos de espera en el mostrador de servicio técnico:", tiempos_espera)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
tiempo_operacion = 12 * 60  # minutos
tasa_llegada = 1 / 15

# Simulación de múltiples días
dias_simulacion = 5
tiempos_llegadas_dias = [np.cumsum(np.random.exponential(1 / tasa_llegada, int(tiempo_operacion / 15))) for _ in range(dias_simulacion)]

# Gráfico de tiempos acumulativos
for i, tiempos_llegada in enumerate(tiempos_llegadas_dias):
    plt.plot(tiempos_llegada, label=f'Día {i+1}')

plt.title("Simulación de Tiempos Acumulados entre Llegadas")
plt.xlabel("Vehículos Llegados")
plt.ylabel("Tiempo (minutos)")
plt.legend()
plt.show()
```

##Conclusion del resultado
En conclusión, el arreglo tiempos_llegada representa los intervalos de tiempo entre las llegadas de vehículos. Esta información es fundamental para analizar y evaluar si la capacidad de carga es adecuada para manejar el flujo diario de vehículos. Al comprender estos intervalos, se pueden identificar posibles cuellos de botella y realizar ajustes necesarios para optimizar la capacidad de carga y mejorar la eficiencia operativa.


#Problema 3:
#Distribución LogNormal
##Contexto del Problema
En el mostrador de servicio técnico de una tienda departamental, los tiempos de espera de los clientes presentan una gran variabilidad, con un promedio de espera de 10 minutos y una desviación estándar de 2 minutos. Este comportamiento se ajusta a una distribución lognormal, útil para describir fenómenos de espera.

##Objetivo Específico
Simular los tiempos de espera de 50 clientes para estudiar el comportamiento de la fila y la duración promedio.

##Distribución Utilizada
La Distribución LogNormal es adecuada para modelar variables cuya transformación logarítmica sigue una distribución normal, como tiempos de espera con una variabilidad que crece en proporción a la media.

```python
clientes = 50
media = 10
desviacion_estandar = 2

tiempos_espera = np.random.lognormal(mean=np.log(media), sigma=desviacion_estandar, size=clientes)
print("Tiempos de espera en el mostrador de servicio técnico:", tiempos_espera)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parámetros
clientes = 50
media = 10
desviacion_estandar = 2

tiempos_espera = np.random.lognormal(mean=np.log(media), sigma=desviacion_estandar, size=clientes)

shape, loc, scale = lognorm.fit(tiempos_espera, floc=0)
x = np.linspace(min(tiempos_espera), max(tiempos_espera), 100)
pdf_fitted = lognorm.pdf(x, shape, loc=loc, scale=scale)

plt.hist(tiempos_espera, bins=15, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.plot(x, pdf_fitted, 'r-', label='Ajuste Lognormal')
plt.title("Distribución de Tiempos de Espera")
plt.xlabel("Tiempo de Espera (min)")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.show()
```

#Conclusion del Resultado
El arreglo tiempos_espera proporciona los tiempos de espera simulados para los clientes. Esta información es valiosa para comprender la duración potencial de las filas en diferentes momentos del día. Con estos datos, se puede optimizar la asignación de personal, reduciendo así los tiempos de espera y mejorando la experiencia del cliente.
